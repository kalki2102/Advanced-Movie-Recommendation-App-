```python
import pandas as pd
data=pd.read_csv('IMDb_All_Genres_etf_clean1.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data=data[['Movie_Title','Director','Actors', 'Rating', 'main_genre', 'side_genre']]
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['genre']=data['side_genre']+" "+ data['main_genre']+""+data['Director']
```

```python
data.drop(['side_genre', 'main_genre'] ,axis= 1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count=CountVectorizer(max_features=5562, stop_words='english')
```

```python
vector=count.fit_transform(data['genre'].values.astype('U')).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vector)
```

```python
dis=sorted(list(enumerate(similar[66])), reverse=True , key= lambda vector:vector[1] )
for i in dis[0:5]:
    print(data.iloc[i[0]].Movie_Title)
dis
```

```python
def recommend(movie_name):
    idx=data[data['Movie_Title']== movie_name].index[0]
    dis= sorted(list(enumerate(similar[idx])) , reverse= True , key= lambda vector:vector[1] )
    for i in dis[1:6]:
        print(data.iloc[i[0]].Movie_Title)
```

```python
recommend('Inception')
```
